```python
plt.style.use("seaborn-white")
alpha= np.linspace(0,1,10000)
plt.figure(figsize=(8,7))
result=[]
for i in alpha:
    if i<10**(-12) or (1-i)<10**(-12):
        result.append(0)
    else:
        result.append(-i*np.log(i)/np.log(2)-(1-i)*np.log(1-i)/np.log(2))
plt.plot(alpha, result,linewidth=1.5,color="firebrick")
#plt.grid(True)
plt.xlim(-0.01,1.01)
plt.ylim(-0.01,1.05)
plt.axvline(x=0.5,ymin=0,ymax=0.95,linestyle="--",color="grey",alpha=0.3)
plt.axhline(y=1, xmin=0.0, xmax=0.5,linestyle="--",color="grey",alpha=0.3)
plt.xlabel(r"$\alpha$",fontsize=16)
plt.ylabel(r"$S\left(\rho_{S}^{\alpha}\right)$",fontsize=16)
plt.text(0.57,1.019, r'$\alpha=0.5$', horizontalalignment='center',verticalalignment='center',fontsize=14,color="navy")
plt.title("Binary Entropy",fontsize=14)
plt.savefig("Binary_entropy.png")
plt.show()
```
