```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
Size_N=100
Size_n=10
Gamma=np.float128(0.5)
Lambda=np.float128(0.5)
```

```python
def alpha(Gamma,Lambda,theta):
    a_0=Lambda
    a_1=np.float128(0.5)*(np.float128(1.0)-Gamma)
    a_2=np.float128(0.5)*(np.float128(1.0)+Gamma)
    return a_0+(a_2+a_1)*np.cos(theta)
def beta(Gamma,Lambda,theta):
    a_1=np.float128(0.5)*(np.float128(1.0)-Gamma)
    a_2=np.float128(0.5)*(np.float128(1.0)+Gamma)
    return (a_1-a_2)*np.sin(theta)
def omega(Gamma,Lambda,theta):
    return np.sqrt(alpha(Gamma,Lambda,theta)**2 + beta(Gamma,Lambda,theta)**2)
def Phi(Gamma,Lambda,theta):
    return np.arctan(beta(Gamma,Lambda,theta)/alpha(Gamma,Lambda,theta))
```

```python
plt.figure(figsize=(8,5))
x=np.linspace(-np.pi,np.pi,100,dtype=np.float128)
plt.plot(x,alpha(Gamma,Lambda,x),label=r"$\alpha(\theta)$",color="navy")
plt.plot(x,beta(Gamma,Lambda,x),label=r"$\beta(\theta)$",color="darkorange")
plt.plot(x,omega(Gamma,Lambda,x),label=r"$\omega(\theta)$",color="forestgreen")
plt.plot(x,Phi(Gamma,Lambda,x),label=r"$\Phi(\theta)$",color="violet",marker="o")
plt.legend()
#plt.savefig("Energias.pdf")
plt.show()
```
