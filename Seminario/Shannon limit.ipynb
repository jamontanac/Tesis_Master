```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def H2(f):
    result=f*np.log(1/f)/np.log(2) + (1-f)*np.log(1/(1-f))/np.log(2) 
    return np.nan_to_num(result)
f=0.1
C= 1-H2(f)
Pb=np.logspace(-15,-1,1000)
R = C/(1- H2(Pb))
```

```python
x=np.linspace(0.3,1,1000)
```

```python
plt.figure(figsize=(9,8))
plt.plot(R,Pb,color="firebrick")
plt.fill_betweenx(np.linspace(10**(-15),1),0.3,1,label="Posible",alpha=0.8 ,facecolor="lightseagreen")
plt.fill_between(R, Pb, interpolate=True,facecolor='salmon',label="Imposible",alpha=1)
plt.yscale("log")
#plt.legend()
plt.xlabel(r"R")
plt.ylabel(r"$P_b$")
plt.xlim(0.3,1)
plt.ylim(10**(-15),1)
plt.text(0.35,10**(-7),"Posible",fontsize=30,rotation=30,family="Tahoma",fontstyle="italic",weight="light")
plt.text(0.7,10**(-9),"Imposible",fontsize=30,rotation=-30,family="Tahoma")
plt.title("")
plt.show()
```

# Bounds

```python
def H2(f):
    result=f*np.log(1/f)/np.log(2) + (1-f)*np.log(1/(1-f))/np.log(2) 
    return np.nan_to_num(result)
```

```python
plt.figure(figsize=(8,8))
delta_gv=np.linspace(0,0.5,1000)
delta=np.linspace(0,1,1000)
plt.plot(delta_gv,1-H2(delta_gv),label="Gilbert-Varshamov")
plt.plot(delta,1-H2(delta/2),label="Hamming")
plt.plot(delta_gv,1-2*delta_gv,label="Plotkin")
plt.legend()
plt.xlabel(r"$\delta$")
plt.ylabel(r"$R$")
plt.axhline(y=0,color="k")
plt.show()
```
